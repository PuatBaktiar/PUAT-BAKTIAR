```python
#soal no 1
import pandas as pd

# Baca dataset
df = pd.read_csv('50_Startups.csv')  # Ganti 'nama_file.csv' dengan nama file dataset Anda

# Identifikasi field yang memiliki data kosong (NaN)
fields_with_nan = df.columns[df.isnull().any()].tolist()

# Isi data kosong dengan nilai mean
for field in fields_with_nan:
    mean_value = df[field].mean()
    df[field].fillna(mean_value, inplace=True)

# Simpan perubahan ke dalam file jika diperlukan
# df.to_csv('nama_file.csv', index=False)  # Ganti 'nama_file.csv' dengan nama file tujuan Anda

# Tampilkan DataFrame setelah pengisian data kosong dengan nilai mean
print(df)
```

```python
#soal no 2
import pandas as pd

# Baca dataset
df = pd.read_csv('50_Startups.csv')  # Ganti 'nama_file.csv' dengan nama file dataset Anda

# Lakukan one-hot encoding pada kolom 'State'
df = pd.get_dummies(df, columns=['State'])

# Tampilkan DataFrame setelah one-hot encoding
print(df)
```

```python
#soal no 4
import pandas as pd

# Baca dataset
df = pd.read_csv('50_Startups.csv')  # Ganti 'nama_file.csv' dengan nama file dataset Anda

# Hitung nilai Tax sesuai rumus
df['Tax'] = (df['Profit'] + df['Marketing Spend'] + df['Administration']) * 0.05

# Tampilkan DataFrame dengan field baru 'Tax'
print(df)
```

```python
#Soal no 5
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Baca dataset
df = pd.read_csv('50_Startups.csv')  # Ganti 'nama_file.csv' dengan nama file dataset Anda

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Tentukan field-field yang ingin di-scaling
fields_to_scale = ['Profit', 'Marketing Spend', 'Administration']  # Sesuaikan dengan field yang Anda inginkan

# Lakukan scaling pada field-field tersebut
df[fields_to_scale] = scaler.fit_transform(df[fields_to_scale])

# Tampilkan DataFrame setelah scaling
print(df)
```
